```python

# Compte the average outcomes for different sample groups at different time periods.zhey
df_trend_0 = ed_sample[ed_sample['ohp_all_ever_matchn_30sep2009_Enrolled']==0]
df_trend_1 = ed_sample[ed_sample['ohp_all_ever_matchn_30sep2009_Enrolled']==1]

Y_00 = df_trend_0['any_visit_pre_ed_Yes'].mean() # pre-treatment, control group
Y_10 = df_trend_1['any_visit_pre_ed_Yes'].mean() # pre-treatment, treatment group
Y_01 = df_trend_0['any_visit_ed_Yes'].mean() # post-treatment, control group
Y_11 = df_trend_1['any_visit_ed_Yes'].mean() # post-treatment, treatment group


# Compute the trend of each individual subject
df_trend_0['diff'] = df_trend_0['any_visit_ed_Yes'] - df_trend_0['any_visit_pre_ed_Yes'] 
df_trend_1['diff'] = df_trend_1['any_visit_ed_Yes'] - df_trend_1['any_visit_pre_ed_Yes']

sd0 = df_trend_0['diff'].std()
sd1 = df_trend_1['diff'].std()

n0 = df_trend_0.shape[0]
n1 = df_trend_1.shape[0]


ATT_hat = (Y_11 - Y_10) - (Y_01 - Y_00)

SE_hat = np.sqrt(sd1**2/n1+sd0**2/n0)
t_statistic = ATT_hat/SE_hat

UCB_95 = ATT_hat+1.96*SE_hat
LCB_95 = ATT_hat-1.96*SE_hat

print("The estimated ATT is %0.4f." %ATT_hat)
print("The estimated standard error of ATT is %0.4f." %SE_hat)
print("The t-statistic is:",t_statistic)
print("The 95% confidence interval for ATT is [",LCB_95,",",UCB_95,"].")
```
